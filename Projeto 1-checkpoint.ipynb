### Filtro da coluna de trabalho infantil e plot de resultados

### Filtro de renda e plot de histogramas por cor/raça

```python
dados2015_renda=dados2015.loc[dados2015.Renda<99999999,['Renda','Cor_ou_Raça']]

fig = plt.figure(figsize=(15,15))
faixa = np.arange(0,10000,1000)
salário = pd.cut(dados2015_renda.Renda,faixa)

plt.subplot(231)
rendabranca = dados2015_renda.Renda[dados2015_renda.Cor_ou_Raça == 'Branca']
plot = rendabranca.plot.hist(bins=faixa,title="Renda dos Brancos 2015", normed = 1,color="lightgrey",alpha=0.8)
plt.xlabel("Renda Mensal")
plt.ylabel("Densidade")
plt.grid(True)
salário.value_counts(sort=False, normalize=True)*100

plt.subplot(232)
rendaparda = dados2015_renda.Renda[dados2015_renda.Cor_ou_Raça == 'Parda']
plot = rendaparda.plot.hist(bins=faixa,title="Renda dos Pardos 2015", normed = 1,color="saddlebrown",alpha=0.8)
plt.xlabel("Renda Mensal")
plt.ylabel("Densidade")
plt.grid(True)
salário.value_counts(sort=False, normalize=True)*100

plt.subplot(233)
rendaindigena = dados2015_renda.Renda[dados2015_renda.Cor_ou_Raça == 'Indígena']
plot = rendaindigena.plot.hist(bins=faixa,title="Renda dos Indígenas 2015", normed = 1,color="red",alpha=0.7)
plt.xlabel("Renda Mensal")
plt.ylabel("Densidade")
plt.grid(True)
salário.value_counts(sort=False, normalize=True)*100

plt.subplot(234)
rendapreta = dados2015_renda.Renda[dados2015_renda.Cor_ou_Raça == 'Preta']
plot = rendapreta.plot.hist(bins=faixa,title="Renda dos Pretos 2015", normed = 1,color="black",alpha=0.8)
plt.xlabel("Renda Mensal")
plt.ylabel("Densidade")
plt.grid(True)
salário.value_counts(sort=False, normalize=True)*100

plt.subplot(235)
rendaamarela = dados2015_renda.Renda[dados2015_renda.Cor_ou_Raça == 'Amarela']
plot = rendaamarela.plot.hist(bins=faixa,title="Renda dos Amarelos 2015", normed = 1,color="yellow",alpha=0.65)
plt.xlabel("Renda Mensal")
plt.ylabel("Densidade")
plt.grid(True)
salário.value_counts(sort=False, normalize=True)*100

plt.tight_layout(pad=0.1, w_pad=3, h_pad=1.0)
```
