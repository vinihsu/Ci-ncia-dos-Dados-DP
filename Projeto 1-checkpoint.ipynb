```python
#Se usar o arquivo descompactado
dados2015=pd.read_csv('PES2015.csv',sep=',')
dados2011=pd.read_csv('PES2011.csv',sep=',')
#Se usar o arquivo compactado como está no dropbox púbico
#pd.read_csv('DOM2015csv.zip', compression='zip', header=0, sep=',', quotechar='"')
```

```python
dados2015=dados2015.loc[dados2015.V4721<999999999,['V6007','V0404','V0701','V8005',"UF",'V0601','V4721']]
#V0404 cor ou raca
#V8005 idade
#V4721 renda
#V0601 sabe ler e escrever
#UF Estado
#V6007nivel de ensino

dados2011=dados2011.loc[dados2011.V4721<999999999,['V6007','V0404','V0701','V8005',"UF",'V0601', 'V4721']]
```

```python
dados2015.V0404 = dados2015.V0404.astype('category')
dados2015.V0404.cat.categories = (['Indígena', 'Branca', 'Preta', 'Amarela', 'Parda', 'Sem Declaração'])

dados2015.UF = dados2015.UF.astype('category')
dados2015.UF.cat.categories = (['RO', 'AC', 'AM', 'RR', 'PA', 'AP', 'TO', 'MA', 'PI', 'CE', 'RN', 'PB', 'PE', 'AL', 'SE', 'BA', 'MG', 'ES', 'RJ', 'SP', 'PR', 'SC', 'RS', 'MS', 'MT', 'GO', 'DF'])

dados2015.loc[(dados2015.V6007==1.0)|(dados2015.V6007==13.0)|(dados2015.V6007==11.0), 'Nível_de_Ensino']='Primário'
dados2015.loc[(dados2015.V6007==6.0)|(dados2015.V6007==4.0), 'Nível_de_Ensino']='1o grau'
dados2015.loc[(dados2015.V6007==5.0)|(dados2015.V6007==7.0)|(dados2015.V6007==3.0)|(dados2015.V6007==2.0), 'Nível_de_Ensino']='2o grau'
dados2015.loc[(dados2015.V6007==8.0)|(dados2015.V6007==9.0), 'Nível_de_Ensino']='Superior'

dados2015.V0601 = dados2015.V0601.astype('category')
dados2015.V0601.cat.categories = ('Sim','Não')

dados2015.V0701 = dados2015.V0701.astype('category')
dados2015.V0701.cat.categories = ('Sim','Não')

dados2011.V0404 = dados2011.V0404.astype('category')
dados2011.V0404.cat.categories = (['Indígena', 'Branca', 'Preta', 'Amarela', 'Parda'])

dados2011.UF = dados2011.UF.astype('category')
dados2011.UF.cat.categories = (['RO', 'AC', 'AM', 'RR', 'PA', 'AP', 'TO', 'MA', 'PI', 'CE', 'RN', 'PB', 'PE', 'AL', 'SE', 'BA', 'MG', 'ES', 'RJ', 'SP', 'PR', 'SC', 'RS', 'MS', 'MT', 'GO', 'DF'])

dados2011.loc[(dados2011.V6007==1.0)|(dados2011.V6007==13.0)|(dados2011.V6007==11.0), 'Nível_de_Ensino']='Fundamental incompleto'
dados2011.loc[(dados2011.V6007==6.0)|(dados2011.V6007==4.0), 'Nível_de_Ensino']='1o grau'
dados2011.loc[(dados2011.V6007==5.0)|(dados2011.V6007==7.0)|(dados2011.V6007==3.0)|(dados2011.V6007==2.0), 'Nível_de_Ensino']='2o grau'
dados2011.loc[(dados2011.V6007==8.0)|(dados2011.V6007==9.0), 'Nível_de_Ensino']='Superior'

dados2011.V0601 = dados2011.V0601.astype('category')
dados2011.V0601.cat.categories = ('Sim','Não')

dados2011.V0701 = dados2011.V0701.astype('category')
dados2011.V0701.cat.categories = ('Sim','Não')
```

```python
dados2015 = dados2015.rename(columns={"V8005": "Idade",'V0701':'Trabalho_Infantil', "V0404":"Cor_ou_Raça", "V0601":"Alfabetizado", "V4721":"Renda", "V6007":"Nível de Ensino" })
dados2011 = dados2011.rename(columns={"V8005": "Idade",'V0701':'Trabalho_Infantil', "V0404":"Cor_ou_Raça", "V0601":"Alfabetizado", "V4721":"Renda", "V6007":"Nível de Ensino" })
```

```python
dados2015_trabalho=dados2015.loc[dados2015.Idade<15,['Trabalho_Infantil','Cor_ou_Raça','Idade']]
dados2015_trabalho = dados2015_trabalho[pd.notnull(dados2015_trabalho["Trabalho_Infantil"])]
ct_trabalho_15=pd.crosstab(dados2015_trabalho.Cor_ou_Raça, dados2015_trabalho.Trabalho_Infantil,normalize='index')*100
ct_trabalho_15
```

```python
dados2015_renda=dados2015.loc[dados2015.Renda<99999999,['Renda','Cor_ou_Raça']]

fig = plt.figure(figsize=(15,15))
faixa = np.arange(0,10000,1000)
salário = pd.cut(dados2015_renda.Renda,faixa)

plt.subplot(231)
rendabranca = dados2015_renda.Renda[dados2015_renda.Cor_ou_Raça == 'Branca']
plot = rendabranca.plot.hist(bins=faixa,title="Renda dos Brancos", normed = 1,color="grey",alpha=0.8)
plt.xlabel("Renda Mensal")
plt.ylabel("Densidade")
plt.grid(True)
salário.value_counts(sort=False, normalize=True)*100

plt.subplot(232)
rendaparda = dados2015_renda.Renda[dados2015_renda.Cor_ou_Raça == 'Parda']
plot = rendaparda.plot.hist(bins=faixa,title="Renda dos Pardos", normed = 1,color="brown",alpha=0.8)
plt.xlabel("Renda Mensal")
plt.ylabel("Densidade")
plt.grid(True)
salário.value_counts(sort=False, normalize=True)*100

plt.subplot(233)
rendaindigena = dados2015_renda.Renda[dados2015_renda.Cor_ou_Raça == 'Indígena']
plot = rendaindigena.plot.hist(bins=faixa,title="Renda dos Indígenas", normed = 1,color="red",alpha=0.5)
plt.xlabel("Renda Mensal")
plt.ylabel("Densidade")
plt.grid(True)
salário.value_counts(sort=False, normalize=True)*100

plt.subplot(234)
rendapreta = dados2015_renda.Renda[dados2015_renda.Cor_ou_Raça == 'Preta']
plot = rendapreta.plot.hist(bins=faixa,title="Renda dos Pretos", normed = 1,color="black",alpha=0.8)
plt.xlabel("Renda Mensal")
plt.ylabel("Densidade")
plt.grid(True)
salário.value_counts(sort=False, normalize=True)*100

plt.subplot(235)
rendaamarela = dados2015_renda.Renda[dados2015_renda.Cor_ou_Raça == 'Amarela']
plot = rendaamarela.plot.hist(bins=faixa,title="Renda dos Pardos", normed = 1,color="yellow",alpha=0.65)
plt.xlabel("Renda Mensal")
plt.ylabel("Densidade")
plt.grid(True)
salário.value_counts(sort=False, normalize=True)*100
```

```python

dados2015_renda.loc[(dados2015.Renda<=1000), 'Classe']='E'
dados2015_renda.loc[(dados2015.Renda<=2000)&(dados2015.Renda>1000), 'Classe']='D'
dados2015_renda.loc[(dados2015.Renda<=5000)&(dados2015.Renda>2000), 'Classe']='C'
dados2015_renda.loc[(dados2015.Renda<=10000)&(dados2015.Renda>5000), 'Classe']='B'
dados2015_renda.loc[(dados2015.Renda>10000), 'Classe']='A'
```

```python
fig = plt.figure(figsize=(12,8))
plt.subplot(233)
classepreta = dados2015_renda.Classe[dados2015_renda.Cor_ou_Raça == 'Preta'].value_counts()
classepreta.plot(kind='pie', autopct="%.2f",legend=False)
plt.title('Pretos')

plt.subplot(232)
classeamarela= dados2015_renda.Classe[dados2015_renda.Cor_ou_Raça == 'Amarela'].value_counts()
classeamarela.plot(kind='pie', autopct="%.2f",legend=False)
plt.title('Amarelos')

plt.subplot(231)
classeindígena= dados2015_renda.Classe[dados2015_renda.Cor_ou_Raça == 'Indígena'].value_counts()
classeindígena.plot(kind='pie', autopct="%.2f",legend=False)
plt.title('Indígenas')

plt.subplot(234)
classebranca = dados2015_renda.Classe[dados2015_renda.Cor_ou_Raça == 'Branca'].value_counts()
classebranca.plot(kind='pie', autopct="%.2f",legend=False)
plt.title('Brancos')

plt.subplot(235)
classeparda = dados2015_renda.Classe[dados2015_renda.Cor_ou_Raça == 'Parda'].value_counts()
classeparda.plot(kind='pie', autopct="%.2f",legend=False)
plt.title('Pardos')
```
